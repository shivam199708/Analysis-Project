```python
# this is called filter method where two condition are give to find the desired result

df_net[(df_net['Category']=='Movie') & (df_net['Year']==2020)]
```

```python
#showing only the title of all the tv shows that were released in India using the filter method 


df_net[(df_net['Category']=='TV Show') & (df_net['Country']=='India')] ['Title']
```

```python
# top 10 director who gave highest number of tv and movie shows

df_net['Director'].value_counts().head(10).plot(kind='bar').set(title='Top 10 Director')
```

```python
# using the filter option along with [or, and] operators

# show all the records where the category is movie and type is  comedie or country is UK 

df_net[(df_net['Category']=='Movie') & (df_net['Type']=='Comedies') | (df_net['Country']=='United Kingdom')]
```

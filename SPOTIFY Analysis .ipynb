```python
df_track.set_index("release_date",inplace = True)
```

```python
df_track.head()
```

```python
df_track.index = pd.to_datetime(df_track.index)
```

```python
df_track.head()
```

```python
df_track[['artists']].iloc[18]
```

```python
df_track["duration"]= df_track["duration_ms"].apply(lambda x:round(x/1000))
df_track.drop("duration_ms",inplace=True,axis=1)
```

```python
df_track.duration.head(5)
```

```python
corr_df=df_track.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("correlation Heatmap Between Variable") 
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df_track.sample(int(0.005*len(df_track)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="danceability", x="popularity", color="c").set(title="danceability Vs Popularity Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title="Loudness Vs Energy Correlation")
```

```python
df_track['dates'] = df_track.index.get_level_values('release_date')
df_track.dates= pd.to_datetime(df_track.dates)
years= df_track.dates.dt.year
```

```python
sns.displot(years, discrete= True, aspect=2, height=7, kind='hist').set(title='Number of Songs Per year')
```

```python
total_dr= df_track.duration
fig_dims= (18,7)
fig,ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax= ax, errwidth= False).set(title='Year Vs Duration')
plt.xticks(rotation=90)
```

```python
total_dr= df_track.duration
sns.set_style(style='whitegrid')
fig_dims= (10,5)
fig,ax = plt.subplots(figsize= fig_dims)
fig = sns.lineplot(x=years, y=total_dr, ax= ax).set(title='Year Vs Duration')
plt.xticks(rotation=60)
```

```python
total_dr= df_track.energy
tl_dr = df_track.loudness
sns.set_style(style='whitegrid')
fig_dims= (10,5)
fig,ax = plt.subplots(figsize= fig_dims)
fig = sns.lineplot(x=tl_dr, y=total_dr, ax= ax).set(title='Loudness Vs Energy')
plt.xticks(rotation=60)
```
